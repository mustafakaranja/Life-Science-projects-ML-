```python
# Point 2 -> data pre processing.
# we are gooing to seperate the feature name and stutus in the diff variable.

X = PD.drop(columns=['name','status'], axis=1)
Y = PD["status"]
```

```python
print(Y)
```

```python
#traing the model , for that we have to spliting data. we have to use the array.

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state =2)

# where as in the train_test_split function test_size we take 20% and for train we take 80%
# X train name are all asign to the y_train having and same as in the test
```

```python
# we have to standardize the data.
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)  # transform the value in the same range.
X_test = scaler.transform(X_test)  # basically we are converting them into the array via standardie lib
```

```python
print(X_train)
```

```python
# training the machine.
# support vector machine model 

#We can use the svm model
```

```python
model = svm.SVC(kernel="linear") # support vector clissifier model
```

```python
#trainng the svm model to with training data
model.fit(X_train, Y_train)
```

```python
#Evalution of our model.

#find the accuracy score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print("accuracy score of training data: ", training_data_accuracy)
```

```python
# test data accuracy score

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print("Accuracy score of training data: ", X_test_prediction)
```
